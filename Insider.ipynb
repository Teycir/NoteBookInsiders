```python
def getClusterInsiderTrades(dataframe,threshold):
    return dataframe.groupby('ticker').filter(lambda x: x['insidername'].nunique() >= threshold)
```

```python
morethan1 = getClusterInsiderTrades(dfRecentPurchases,2)
morethan1
```

```python
def getTickers(df):
    uniquetickers = df['ticker'].unique()
    listp = ' '.join(uniquetickers)
    tickers = []
    for twords in listp.split():
        tickers.append(twords)
    print(' '.join(tickers))
    return tickers
    
```

```python
def getTickersLimit(df,limit):
    uniquetickers = df['ticker'].unique()
    listp = ' '.join(uniquetickers)
    tickers = []
    for twords in listp.split():
        tickers.append(twords)
    print(' '.join(tickers[0:limit]))
    return tickers[0:limit]
```

```python
def getTickersDollar(df):
    uniquetickers = df['ticker'].unique()
    listp = ' '.join(uniquetickers)
    tickers = []
    for twords in listp.split():
        twords="$"+twords
        tickers.append(twords)
    print(' '.join(tickers))
    return tickers
    
```

```python
def getTickersDollarLimit(df,limit):
    uniquetickers = df['ticker'].unique()
    listp = ' '.join(uniquetickers)
    tickers = []
    for twords in listp.split():
        twords="$"+twords
        tickers.append(twords)
    print(' '.join(tickers[0:limit]))
    return tickers[0:limit]
    
```

```python
def getTickers2formats(df,limit):
    getTickersDollarLimit(df,limit)
    getTickersLimit(df,limit)
```

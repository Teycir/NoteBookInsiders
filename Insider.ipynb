```python
def getTickers(df):
    uniquetickers = df['ticker'].unique()
    listp = ' '.join(uniquetickers)
    tickers = []
    for twords in listp.split():
        tickers.append(twords)
    print(' '.join(tickers))
    return tickers
    
```

```python
def getTickersLimit(df,limit):
    uniquetickers = df['ticker'].unique()
    listp = ' '.join(uniquetickers)
    tickers = []
    for twords in listp.split():
        tickers.append(twords)
    print(' '.join(tickers[0:limit]))
    return tickers[0:limit]
```

```python
def getTickersDollar(df):
    uniquetickers = df['ticker'].unique()
    listp = ' '.join(uniquetickers)
    tickers = []
    for twords in listp.split():
        twords="$"+twords
        tickers.append(twords)
    print(' '.join(tickers))
    return tickers
    
```

```python
def getTickersDollarLimit(df,limit):
    uniquetickers = df['ticker'].unique()
    listp = ' '.join(uniquetickers)
    tickers = []
    for twords in listp.split():
        twords="$"+twords
        tickers.append(twords)
    print(' '.join(tickers[0:limit]))
    return tickers[0:limit]
    
```

```python
def getTickers2formats(df,limit):
    getTickersDollarLimit(df,limit)
    getTickersLimit(df,limit)
```

```python
def getTopOfficerPurchases(df,title,limit):
    #df[df.sum(axis=1).str.contains('-->')]
    topOfficerPurchases = df[df['title'].str.contains(title)]
    topOfficerPurchasesShort= topOfficerPurchases[['ticker','industry','country','date_y','title', 'tradetype','price_y','value','tradedate','insidername']]
    topOfficerPurchasesShortSorted = topOfficerPurchasesShort.sort_values(['value'], ascending=[False])
    print(getTickers2formats(topOfficerPurchasesShortSorted,limit))
    return topOfficerPurchasesShortSorted.head(limit)
```
